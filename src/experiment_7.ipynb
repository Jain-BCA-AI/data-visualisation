## HeatMap and Wordcloud

```python
import seaborn as sns
import numpy as np
```

```python
#Loading the dataset 
tips=sns.load_dataset("tips")
tips.head()
```

```python
#Finding the correlation
tips.corr(numeric_only=True)
```

```python
#Generating the heatmap for correlation for tips
sns.heatmap(tips.corr(numeric_only=True))
```

```python
#Adding cmap parameter to heatmap
sns.heatmap(tips.corr(numeric_only=True),cmap='coolwarm')
```

```python
#Adding annot parameter to heatmap
sns.heatmap(tips.corr(numeric_only=True),cmap='Greens',annot=True)
```

```python
#Adding linecolor,linewidth parameters to heatmap
sns.heatmap(tips.corr(numeric_only=True),cmap='PiYG',annot=True,linecolor='black',linewidth=3)
```

## Wordcloud

```python
#Importing packages
from wordcloud import WordCloud,STOPWORDS
import wikipedia as wiki
from PIL import Image
```

```python
#Setting stopwords 
stop_w=set(STOPWORDS)
```

```python
#Getting information from wikipedia
info=wiki.summary('python programming')
info
```

```python
#Generating the worldcloud
wcl=WordCloud(stopwords=stop_w).generate(info)
```

```python
#Displaying the wordcloud
img=wcl.to_image()
img
```

```python
#Getting information from wikipedia
out=wiki.summary('Jain University',sentences=5)
out
```

```python
import os
dir = os.getcwd()
filename = dir + "/../assets/image/bird.png"
print(filename)
```

```python
#Masking the image
mask_img=np.array(Image.open(filename))
```

```python
#generating the wordcloud in shape of masked image
wc=WordCloud(mask=mask_img,
             prefer_horizontal=0.5,
             background_color="rgba(255, 255, 255, 0)", 
             mode="RGBA",
             stopwords=stop_w).generate(out)
wc
```

```python
img=wc.to_image()
```
