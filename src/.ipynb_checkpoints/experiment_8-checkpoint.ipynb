## Histogram and Pie Chart

```python
#Importing libraries
import matplotlib.pyplot as plt
```

```python
#Creating a list which contains different values of range
age = [24,23,58,75,81,45,65,25,41,32,25,41,14,78,90,12,
       25,34,39,58,28,17,45,65,78,45,56,26,16,1,5,6,9,8,
       5,2,2,45,65,25,87,79,45,46,19,25,45,63,12,85]
```

```python
l=len(age)
mi=min(age)
ma=max(age)
print("length of age:{0}\nmax-value of age:{1}\nmin-value of age:{2}".format(l,mi,ma))
```

```python
#Plotting histogram
plt.hist(age)
plt.show()
```

```python
#Plotting histogram with labels
plt.figure(figsize=(4,4))
plt.xlabel('Ages')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.hist(age)
plt.show()
```

```python
#Adding edge color to the plot
plt.figure(figsize=(4,4))
plt.xlabel('Ages')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.hist(age,ec='red')
plt.show()
```

```python
#Changing color of the bars
plt.figure(figsize=(4,4))
plt.xlabel('Ages')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.hist(age,color='brown',ec='black')
plt.show()
```

```python
#Setting the range value of x-axis
y=[1,20,40,60,80,90]
plt.figure(figsize=(4,4))
plt.xlabel('Ages')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.hist(age,color='brown',ec='black')
plt.xticks(y)
plt.show()
```

```python
#Setting the number of bins
plt.figure(figsize=(4,4))
plt.xlabel('Ages')
plt.ylabel('Count')                                  
plt.title('Age Distribution')
plt.hist(age,color='brown',bins=5,ec='black')
plt.show()
```

```python
#width=(max-min)/bin
width=(max(age)-min(age))/5
print(width)
```

```python
#Adding legend
plt.figure(figsize=(4,4))
plt.xlabel('Ages')
plt.ylabel('Count')                                  
plt.title('Age Distribution')
plt.hist(age,color='brown',bins=5,ec='black')
plt.legend(['age'])
plt.show()
```

## Piechart

```python
#Creating two list 
stu_grade=['Excellent','Good','Average','Poor']
stu_count=[35,15,30,20]
```

```python
#Plotting pie chart
plt.pie(stu_count);
```

```python
#Adding labels
plt.figure(figsize=(3,3))
plt.pie(stu_count,labels=stu_grade);
```

```python
#Changing the starting angle of pie chart
plt.figure(figsize=(3,3))
plt.pie(stu_count,labels=stu_grade,startangle=90); #pie chart always works in anticlock wise
```

```python
#Exploding the pie chart
myexp=[0,0,0,0.2]
plt.figure(figsize=(3,3))
plt.pie(stu_count,labels=stu_grade,startangle=90,explode=myexp);
```

```python
#Adding shadow to chart
plt.figure(figsize=(3,3))
plt.pie(stu_count,labels=stu_grade,startangle=90,explode=myexp,shadow=True);
```

```python
#Adding customized colors 
co=['green','orange','yellow','red']
plt.figure(figsize=(3,3))
plt.pie(stu_count,labels=stu_grade,startangle=90,explode=myexp,shadow=True,colors=co);
```

```python
#Adding percentage of values in the plot
plt.figure(figsize=(3,3))
plt.pie(stu_count,labels=stu_grade,startangle=90,explode=myexp,shadow=True,colors=co,autopct='%2.2f%%');
```

```python
#Adding legend
plt.figure(figsize=(5,4))
plt.pie(stu_count,labels=stu_grade,startangle=90,explode=myexp,shadow=True,colors=co,autopct='%2.2f%%');
plt.legend()
```

```python
#Adding title to legend and aligning the legend box
plt.figure(figsize=(4,4))
plt.pie(stu_count,labels=stu_grade,startangle=90,explode=myexp,shadow=True,colors=co,autopct='%2.2f%%');
plt.legend(loc='center right',bbox_to_anchor=(1, 0, 0.5, 1),title='Student Performance')
```

```python
#Adding frame and rotating axis of labels
plt.figure(figsize=(5,6))
plt.pie(stu_count,labels=stu_grade,startangle=90,explode=myexp,
        colors=co,autopct='%2.2f%%',rotatelabels=True,frame=True);
plt.legend(loc='center right',bbox_to_anchor=(1, 0, 0.5, 1),title='Student Performance')
plt.show()
```
